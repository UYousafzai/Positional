```python
import numpy as np
from pandas import DataFrame, read_csv
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_feather('complete.feather')
dataset.drop(['special','ID','name','full_name','club','club_logo','league','birth_date','real_face', 'flag','nationality','photo','eur_value','eur_wage','eur_release_clause','overall','potential','pac','sho','pas','dri','def','phy'], axis=1,inplace=True)
dataset.drop(dataset.columns[[x for x in range (135,162)]], axis=1,inplace=True) #71 for just excluding traits fields.

display(dataset.head())
```

```python
def nonlin(x,deriv=False):
    if(deriv==True):
        return x*(1-x)
    return 1/(1+np.exp(-x))

class ANN:
    def __init__(self):
        self.layers=5 #including input and output
        self.units = [51,80,50,40,1]#27]
        
        # seed random numbers to make calculation
        # deterministic (just a good practice)
        np.random.seed(1)
        self.alpha = 0.00001
        
        # initialize weights randomly with mean 0
        self.syn0 = 2*np.random.random((self.units[0]+1,self.units[1])) - 1
        self.syn1 = 2*np.random.random((self.units[1]+1,self.units[2])) - 1
        self.syn2 = 2*np.random.random((self.units[2]+1,self.units[3])) - 1
        self.syn3 = 2*np.random.random((self.units[3]+1,self.units[4])) - 1

    
    def forward(self, X):
        m = len(X)
        self.l0 = np.ones((m,len(self.syn0)))         # Add bias column
        self.l0[:,1:] = X

        self.l1 = np.ones((m,len(self.syn1)))         # Add bias column
        self.l1[:,1:] = nonlin(np.dot(self.l0,self.syn0))

        self.l2 = np.ones((m,len(self.syn2)))         # Add bias column
        self.l2[:,1:] = nonlin(np.dot(self.l1,self.syn1))

        self.l3 = np.ones((m,len(self.syn3)))         # Add bias column
        self.l3[:,1:] = nonlin(np.dot(self.l2,self.syn2))

        self.l4 = np.dot(self.l3,self.syn3)
        return self.l4
    
    def backward(self, y):
        # how much did we miss?
        self.l4_error = (y - self.l4)
        np.nan_to_num(self.l4_error,copy=False)

        # in what direction is the target value?
        # were we really sure? if so, don't change too much.
        self.l4_delta = self.l4_error
        # how much did each l2 value contribute to the l2 error (according to the weights)?
        self.l3_error = self.l4_delta.dot(self.syn3.T)
        self.l3_delta = self.l3_error* nonlin(self.l3,deriv=True)
        self.l2_error = self.l3_delta[:,1:].dot(self.syn2.T)
        self.l2_delta = self.l2_error* nonlin(self.l2,deriv=True)
        self.l1_error = self.l2_delta[:,1:].dot(self.syn1.T)
        self.l1_delta = self.l1_error* nonlin(self.l1,deriv=True)
        
        self.syn3 += self.l3.T.dot(self.l4_delta) * self.alpha
        self.syn2 += self.l2.T.dot(self.l3_delta[:,1:]) * self.alpha
        self.syn1 += self.l1.T.dot(self.l2_delta[:,1:]) * self.alpha
        self.syn0 += self.l0.T.dot(self.l1_delta[:,1:]) * self.alpha
    
ann = ANN()
for iter in range(50000):
    ann.forward(X_train)
    ann.backward(y_train)
    
    if (iter% 100) == 0:
        print("Error:" + str(np.mean(np.abs(ann.l4_error))))
        print(ann.l4.shape)
```
